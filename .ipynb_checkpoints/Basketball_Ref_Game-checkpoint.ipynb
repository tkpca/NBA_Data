```python
from bs4 import BeautifulSoup as bs

import requests
import re
import numpy as np
from datetime import datetime

url = 'https://www.basketball-reference.com/boxscores/201604010CHO.html'

game_list_season=[]

r  = requests.get(url)
data = r.text
soup = bs(data)
```

```python
date_temp = soup.find_all(['h1'])
date_temp=date_temp[0].text
date1,date2,date3=(date_temp.split(','))
date2=date2.strip()
date3=date3.strip()
game_dt=date2 +', '+date3
print(game_dt)

headings = soup.find_all(['h2'])

home_tm=(headings[4].text)[:(headings[4].text).find('(')].rstrip() #remove trailing spaces
away_tm=(headings[12].text)[:(headings[12].text).find('(')].rstrip()
print(home_tm)
print(away_tm)
```

```python
#use pandas to pull the relevant html tables
import pandas as pd

html = url
tables = pd.read_html(html)
```

```python
team_dict = pd.read_csv('xRef_teams.csv',index_col='Tm_Long') #use this to convert the team names to abbreviations
team_map=(team_dict.to_dict())['Tm_Abb']
home_tm=team_map[home_tm] #convert to home abbreviation
away_tm=team_map[away_tm] #convert to away abbreviation

date_temp = datetime.strptime(date2 +' '+ date3, '%B %d %Y')
gid=str(date_temp.year)+str(date_temp.month).zfill(2)+str(date_temp.day).zfill(2)+'_'+home_tm+'_'+away_tm #make a key for the game, year-month-day + home and away teams
```

```python
df_hm=tables[0] #home team table Q1 · Q2 · H1 · Q3 · Q4 · H2 map to tables 1,2,3,4,5,6,7 respectively
df_aw=tables[8] #away team table
```

```python
#need to remove the multindex
df_hm.columns = df_hm.columns.droplevel(0)
df_aw.columns = df_aw.columns.droplevel(0)
```

```python
df_hm['hm_away']='H'
df_hm['team']=home_tm
df_hm['gm_date']=game_dt
df_hm=df_hm[~df_hm['Starters'].isin(['Reserves','Team Totals'])] #remove totals and reserve junk columns

df_aw['hm_away']='A'
df_aw['team']=away_tm
df_aw['gm_date']=game_dt
df_aw=df_aw[~df_aw['Starters'].isin(['Reserves','Team Totals'])] #remove totals and reserve junk columns
```

```python
df_hm.head(5)
```

```python
box_df=df_aw.append(df_hm,ignore_index=False) #merge home and away dateframes
box_df.columns = map(str.lower, box_df.columns)
```

```python
box_df['status'] = ['I' if ("Dress") in x or ('Team') in x or ('With') in x or ('Not') in x else 'A' for x in box_df['mp']]  #flag active vs inactive players
box_df.head()
```

```python
box_df['season']=date3+'-'+str(int(date3)+1) #add season tag
box_df['min'], box_df['sec'] = (box_df['mp'].str.split(':', 1).str) #split min and seconds
box_df['gid']=gid
box_df.head()



                     
```

```python
box_df.columns
```

```python
box_df = box_df.rename(columns={'starters': 'full_name','+/-':'pm','trb':'reb','orb':'oreb','drb':'dreb'})
```

```python
box_df.drop('mp'.split(' '),axis=1) #remove redundant columns
box_df.head()
```

```python
col_list=list(box_df)
col_list.sort()
col_list
```
