```python
import datetime
import numpy as np
import pandas as pd
import random
import time
import json
import requests

pd.options.display.float_format = '{:,.2f}'.format
pd.set_option('display.max_columns', None)
```

```python
'''This stores the default variables with the query dictionaries for the default season of 2019-20'''
payload={
'LastNGames': '0',
'LeagueID':'00',
'Month':'0',
'OpponentTeamID': '0',
'PORound': '0',
'PerMode': 'PerGame',
'Period': '0',
'Season': '2010-11',
'TeamID': '0',
'SeasonType':"Regular Season"
}

headers = {
    'Host': 'stats.nba.com',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:71.0) Gecko/20100101 Firefox/71.0',
    'Accept': 'application/json, text/plain, */*',
    'Accept-Language': 'en-US,en;q=0.5',
    'Referer': 'https://stats.nba.com/',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
        }
```

```python
def pull_data(s,payload,headers):
    url='https://stats.nba.com/stats/leaguedashplayerbiostats?'
    with s.get(url,headers=headers,params=payload) as response:
        try:
            assert response.status_code==200
            ret=response.json()
            return ret
        except AssertionError:
            print(response)
    
    
```

```python
'''main body of the program
first determine what years to pull in the loop
second loop thorough those years and return the response with the player data in a dictionary
'''
#first_year=int(input('What is the first year of the season to pull:'))
#last_year=int(input('What is the last year of the season to pull:'))
first_year=int(input('Pick the first season to scrape:'))
last_year=int(input('Pick the last season to scrape:'))
#wait_base=int(input('~how many seconds should the requests take in total?'))
years_to_pull=[x for x in range(first_year,last_year+1)] #create a list with the years to loop through

s= requests.Session()
cnt=0
results={}
for x in years_to_pull:
    season_param=((str(x)+'-'+str((int(x)+1))[2:4]))
    payload['Season']=season_param
    print(f'Currently processing season: {season_param}')
    response= pull_data(s,payload,headers)
    results[cnt]=response
    cnt+=1
```

```python
#setup the column headings for the dataframes
col_headings=results[0]['resultSets'][0]['headers']
#create a dictionary with the season as the key and a list of results as  the value'
data={}
data={results[i]['parameters']['Season']:results[i]['resultSets'][0]['rowSet'] for i in range (len(results))}
```

```python
'''build a dataframe with player data for each year'''

df_player_bio=pd.DataFrame()

for k,v in data.items():
    df_temp=pd.DataFrame(v,columns=col_headings)
    df_temp['SEASON']=k
    df_player_bio=df_player_bio.append(df_temp)
    
    
```

```python
df_player_bio.to_hdf(f'df_player_bio_data.h5', key='df_player_bio', mode='w',complevel =9,complib='bzip2')
```
